```python
# 4. Advogar ou ser representante de vendas faz você dormir menos?
    # 1. Filtrar o DataFrame para apenas as linhas onde a profissão é 'Advogado' ou 'Representante de Vendas'
    selected = df[df['Profissão'].isin(['Advogado(a)', 'Representante de Vendas'])]

    # 2. Calcular a média das horas de sono para o grupo filtrado
    mean_sleep_selected = selected['Duração do sono'].mean()

    # 3. Calcular a média geral das horas de sono do DataFrame completo
    mean_sleep_overall = df['Duração do sono'].mean()

    # 4. Imprimir as médias para comparação
    print("Média de horas de sono para Advogado/Representante de Vendas: {:.2f}".format(mean_sleep_selected))
    print("Média geral de horas de sono: {:.2f}".format(mean_sleep_overall))

    # 5. Verificar se o grupo de Advogado/Representante de Vendas dorme menos que a média geral
    if mean_sleep_selected < mean_sleep_overall:
        print("Conclusão: Advogar ou ser representante de vendas está associado a menos horas de sono.")
    else:
        print("Conclusão: Advogar ou ser representante de vendas não está associado a uma redução no sono.")
```

```python
# 5. Entre quem fez Enfermagem e quem fez Medicina, quem tem menos horas de sono?
selected_health = df[df['Profissão'].isin(['Enfermeiro(a)', 'Médico(a)'])]
mean_sleep_health = selected_health.groupby('Profissão')['Duração do sono'].mean()
print("\n4) Média de horas de sono por profissão (Enfermagem e Medicina):")
print(mean_sleep_health)
if mean_sleep_health['Enfermeiro(a)'] < mean_sleep_health['Médico(a)']:
    print("Conclusão: Quem fez Enfermagem tem, em média, menos horas de sono.")
elif mean_sleep_health['Médico(a)'] < mean_sleep_health['Enfermeiro(a)']:
    print("Conclusão: Quem fez Medicina tem, em média, menos horas de sono.")
else:
    print("Conclusão: As médias de horas de sono são iguais para Enfermagem e Medicina.")
```

```python
# 7. Descobrir qual a profissão menos frequente no conjunto:

# Conta quantas vezes cada profissão aparece
prof_counts = df['Profissão'].value_counts()

# Identifica qual profissão tem a menor contagem
less_frequent_prof = prof_counts.idxmin()

# Imprime o resultado
print("\n6) Profissão menos frequente:")
print("Profissão: {}, Frequência: {}".format(less_frequent_prof, prof_counts.min()))
```

```python
# 8. Quem tem maior pressão sanguínea média, homens ou mulheres?

# 1. Extrair a parte sistólica da pressão sanguínea
# Aqui, usamos .str.split('/') para separar a string e pegamos o primeiro valor.
df['Sistolica'] = df['Pressão sanguínea'].str.split('/').str[0]

# 2. Converter a coluna extraída para valores numéricos
# errors='coerce' fará com que valores que não possam ser convertidos se tornem NaN.
df['Sistolica'] = pd.to_numeric(df['Sistolica'], errors='coerce')

# 3. Calcular a pressão sistólica média por gênero
bp_by_gender = df.groupby('Gênero')['Sistolica'].mean()

# 4. Exibir o resultado
print("Pressão sistólica média por gênero:")
print(bp_by_gender)
```

```python
# 9. É predominante entre os participantes dormir 8 horas por dia (usando a Moda):
sleep_mode = df['Duração do sono'].mode()
print("\n8) Moda da duração do sono: ", sleep_mode.tolist())
if 8 in sleep_mode.values:
    print("Conclusão: Dormir 8 horas é a moda entre os participantes.")
else:
    print("Conclusão: Dormir 8 horas não é a moda predominante.")
```

```python
# Ensure the correct column name for heart rate and steps
heart_rate_col = 'Frequência cardíaca'  # Adjust this based on your dataset
steps_col = 'Passos diários'  # Adjust this based on your dataset

# Split the dataset
high_hr = df[df[heart_rate_col] > 70][steps_col].mean()
low_hr = df[df[heart_rate_col] <= 70][steps_col].mean()

# Compare the means
print(f'Média de passos diários para FC > 70: {high_hr:.2f}')
print(f'Média de passos diários para FC <= 70: {low_hr:.2f}')

if high_hr > low_hr:
    print("Pessoas com FC > 70 dão mais passos diários, em média.")
else:
    print("Pessoas com FC <= 70 dão mais passos diários, em média.")
```
